```python
def get_embedding(text):
    return transformer.encode(text).tolist()
```

```python
def generate_paraphrases(format_function, data, tokenizer, model, device, combination, similarity_query, additional_args=None,
                         few_shots=None, max_new_tokens=100, temperature=0.2):
    paraphrased_sentences = []

    for sentence in data['Sentence 1'][:100]:
        if additional_args:
            formatted_prompt = format_function(sentence, combination, similarity_query, few_shots)
        else:
            formatted_prompt = format_function(sentence, combination, similarity_query)

        print(formatted_prompt)

        inputs = tokenizer(formatted_prompt, return_tensors="pt")
        input_ids = inputs['input_ids'].to(device)
        attention_mask = inputs['attention_mask'].to(device)

        output = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_new_tokens=max_new_tokens,
                                        temperature=temperature, do_sample=True)

        generated_text = output[:, input_ids.shape[-1]:]
        response = tokenizer.decode(generated_text[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
        print(f"Unprocessed response: {response}")
        response = process_output(response)
        print(f"Processed response: {response}")
        paraphrased_sentences.append(response)

    return paraphrased_sentences
```

```python
def process_output(response):
    response = re.sub(r'["\']', '', response)

    match = re.search(r'<ANSWER>(.*?)</ANSWER>', response, re.DOTALL)
    if match:
        return match.group(1).strip()

    return response.strip()
```

```python
'''
За контекстот додаден во промптот, земаме embedding за реченицата и врз основа на истиот бараме слични зборови во граф базата.
На овој начин, зборовите коишто ќе се добијат ќе бидат блиски до значењето и контекстот што го дава целата реченица, а не само на секој збор одделно.
Во промптот се додава и „дефиниција" за значењето на различните зборовни групи како синоними, хиперними, хипоними и слично со цел моделот да разликува меѓу истите.
'''
def format_prompt_with_context(sentence, combination, similarity_query, few_shots=None):
    prompt = f"Original sentence: '{sentence}'\n"

    if few_shots:
        prompt += "Here are some example paraphrases to guide you:\n"
        for idx, (original, paraphrased) in enumerate(few_shots, start=1):
            prompt += (
                f"{idx}. Example: "
                f"Original sentence: '{original}', "
                f"Paraphrased sentence: <ANSWER>{paraphrased}</ANSWER>\n"
            )

    sentence_embedding = get_embedding(sentence)
    similarity_results = similarity_query(sentence_embedding, combination)

    prompt += "\nHere are some relevant words and their meanings based on the similarity:\n"

    relation_explanations = {
        'SYNONYM': 'Synonyms are words with the same or similar meanings as the original word.',
        'HYPERNYM': 'Hypernyms are more general terms that describe a category the original word belongs to.',
        'HYPONYM': 'Hyponyms are more specific terms that describe a subset of the original word.',
        'ANTONYM': 'Antonyms are words with the opposite meaning of the original word.',
        'MERONYM': 'Meronyms are words that represent a part of the whole described by the original word.',
        'EXAMPLE': 'Examples show how the word is typically used in sentences.'
    }

    explained_relations = set()

    for record in similarity_results:
        word = record['synset_name']
        relation = record['relations'][0]
        synset_definition = record["synset_definition"]

        if relation not in explained_relations:
            prompt += f"\n- **{relation_explanations.get(relation, 'Relation explanation not available')}**\n"
            explained_relations.add(relation)

        if relation == 'EXAMPLE':
            if record['examples']:
                prompt += f"  Example usage of '{word}': {record['examples'][0]}\n"
        else:
            prompt += f"  '{word}': {synset_definition}\n"

    prompt += (
        "\nNow, paraphrase the original sentence using the provided words and their definitions. "
        "Ensure that the meaning of the sentence remains the same, but feel free to use different wordings or phrases where appropriate. "
        "The paraphrased sentence should reflect the same idea as the original, using relevant terms from the list above. "
        "Enclose your paraphrased sentence within <ANSWER></ANSWER> tags.\n"
    )

    return prompt
```

```python
def get_few_shots_array(dataset, number):
  number = min(number, len(dataset))

  random_indices = random.sample(range(len(dataset)), number)

  few_shots = []
  for idx in random_indices:
      original_sentence = dataset['Sentence 1'].iloc[idx]
      paraphrased_sentence = dataset['Sentence 2'].iloc[idx]
      few_shots.append((original_sentence, paraphrased_sentence))

  return few_shots
```

```python
from evaluate import load

def evaluate_model(predictions, references, experiment_name, lang='en'):
    print(f" Evaluating Experiment: {experiment_name}")

    bertscore_metric = load('bertscore')
    bertscore_result = bertscore_metric.compute(predictions=predictions, references=references, lang=lang)
    mean_precision = np.mean(bertscore_result['precision'])
    mean_recall = np.mean(bertscore_result['recall'])
    mean_f1 = np.mean(bertscore_result['f1'])

    print("\n BERTScore Metrics ")
    print(f"Precision: {mean_precision} | Recall: {mean_recall} | F1: {mean_f1}")

    rouge_metric = load('rouge')
    rouge_result = rouge_metric.compute(predictions=predictions, references=references)

    print("\n ROUGE Scores ")
    print("\n".join([f"{key}: {value}" for key, value in rouge_result.items()]))

    bleu_metric = load('bleu')
    bleu_result = bleu_metric.compute(predictions=predictions, references=references)

    print("\n BLEU Score ")
    print("\n".join([f"{key}: {value}" for key, value in bleu_result.items()]))

    meteor_metric = load('meteor')
    meteor_result = meteor_metric.compute(predictions=predictions, references=references)

    print("\n METEOR Score ")
    print(f"meteor: {meteor_result['meteor']}")
```

```python
'''
Функција за извршување на Cypher query.
'''
def run_query(query, parameters=None):
    with driver.session() as session:
        result = session.run(query, parameters)
        return [record.data() for record in result]
```

```python
'''
Во ова query пребаруваме слични synsets во Neo4j според даден ембединг на целата реченица којашто треба да се парафразира,
користејќи косинусна сличност, при што враќаме информации за сличноста, дефиницијата на synsetот и сите типови релации кои го поврзуваат пронајдениот synset со други јазли.
Се враќаат само 5-те најслични јазли се со цел да не го преоптоваруваме моделот со голем промпт.
Беше испробана и варијантата со лимитирање на 10 јазли, но сепак немаше подобрување во резултатите.
'''
def run_similarity_query_wo_examples(query_embedding, relations):
  query = f"""
    WITH $query_embedding AS query_embedding
    MATCH (synset1:Synset)-[r]->(synset2:Synset)
    WHERE type(r) IN $relations
    AND synset1.embedding IS NOT NULL AND synset2.embedding IS NOT NULL

    WITH DISTINCT synset2,
        COLLECT(DISTINCT type(r)) AS relations,  // Collect all relations for each synset
        reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (query_embedding[i] * synset2.embedding[i])) AS dot_product,
        reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (query_embedding[i] * query_embedding[i])) AS norm_query,
        reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (synset2.embedding[i] * synset2.embedding[i])) AS norm_synset2

    WITH synset2, dot_product / (sqrt(norm_query) * sqrt(norm_synset2)) AS similarity, relations

    RETURN DISTINCT synset2.name AS synset_name,
                    synset2.definition AS synset_definition,
                    similarity,
                    relations  // Now returns a list of relations
    ORDER BY similarity DESC
    LIMIT 5
  """

  parameters = {
      "query_embedding": query_embedding,
      "relations": relations
  }

  results = run_query(query, parameters)
  return results
```

```python
'''
Ова query се разликува од претходното по тоа што тука се вклучуваат и examples јазлите.
'''
def run_similarity_query_w_examples(query_embedding, relations):
    query = f"""
      WITH $query_embedding AS query_embedding

      MATCH (synset1:Synset)-[r]->(synset2:Synset)
      WHERE type(r) IN $relations
      AND synset1.embedding IS NOT NULL AND synset2.embedding IS NOT NULL

      WITH DISTINCT synset2,
          type(r) AS relation,
          reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (query_embedding[i] * synset2.embedding[i])) AS dot_product,
          reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (query_embedding[i] * query_embedding[i])) AS norm_query,
          reduce(sum = 0, i IN range(0, size(query_embedding)-1) | sum + (synset2.embedding[i] * synset2.embedding[i])) AS norm_synset2

      WHERE norm_query > 0 AND norm_synset2 > 0

      WITH synset2, dot_product / sqrt(norm_query * norm_synset2) AS similarity, relation
      ORDER BY similarity DESC
      LIMIT 10

      OPTIONAL MATCH (synset2)<-[:EXAMPLE]-(example:Example)

      WITH synset2, similarity,
           [x IN COLLECT(relation) WHERE x IS NOT NULL] AS relations,
           [x IN COLLECT(example.text) WHERE x IS NOT NULL][..5] AS examples

      RETURN DISTINCT
          synset2.name AS synset_name,
          synset2.definition AS synset_definition,
          similarity,
          relations,
          examples
      ORDER BY similarity DESC
      LIMIT 5
    """

    parameters = {
        "query_embedding": query_embedding,
        "relations": relations
    }

    results = run_query(query, parameters)
    return results
```

Код користен за конструкција на граф базата

```python
def create_relationship(tx, synset1, synset2, rel_type):
    query = f"""
    MATCH (a:Synset {{id: $id1}}), (b:Synset {{id: $id2}})
    MERGE (a)-[:{rel_type}]->(b)
    """
    tx.run(query, id1=synset1.name(), id2=synset2.name())
```

```python
def create_synset(tx, synset):
    definition = synset.definition()
    embedding = get_embedding(definition)

    query = """
    MERGE (s:Synset {id: $id, name: $name, pos: $pos})
    SET s.definition = $definition, s.embedding = $embedding
    """
    tx.run(query, id=synset.name(), name=synset.lemmas()[0].name(), pos=synset.pos(),
           definition=definition, embedding=embedding)
```

```python
def process_synset_batch(batch):
    synsets_data = []
    relationships_data = []
    examples_data = []

    for synset in batch:

        synsets_data.append({
            "id": synset.name(),
            "name": synset.lemmas()[0].name(),
            "pos": synset.pos(),
            "definition": synset.definition(),
            "embedding": get_embedding(synset.definition())
        })

        for related_synset in wn.synsets(synset.lemmas()[0].name()):
            relationships_data.append((synset.name(), related_synset.name(), "SYNONYM"))

        for hypernym in synset.hypernyms():
            relationships_data.append((synset.name(), hypernym.name(), "HYPERNYM"))

        for hyponym in synset.hyponyms():
            relationships_data.append((synset.name(), hyponym.name(), "HYPONYM"))

        for meronym in synset.part_meronyms():
            relationships_data.append((synset.name(), meronym.name(), "MERONYM"))

        for lemma in synset.lemmas():
            for antonym in lemma.antonyms():
                relationships_data.append((synset.name(), antonym.synset().name(), "ANTONYM"))

        for example in synset.examples():
            examples_data.append({"text": example, "synset_id": synset.name()})

    with driver.session() as session:
        session.write_transaction(batch_insert, synsets_data, relationships_data, examples_data)

    global processed_count
    processed_count += len(batch)
    if processed_count % 100 == 0:
        print(f"Processed {processed_count} synsets of {len(all_synsets)}")
```

```python
def batch_insert(tx, synsets_data, relationships_data, examples_data):
    """Inserts nodes and relationships in a single transaction"""

    query_synsets = """
    UNWIND $synsets AS synset
    MERGE (s:Synset {id: synset.id})
    SET s.name = synset.name, s.pos = synset.pos,
        s.definition = synset.definition, s.embedding = synset.embedding
    """
    tx.run(query_synsets, synsets=synsets_data)

    query_relationships = """
    UNWIND $relationships AS rel
    MATCH (s1:Synset {id: rel[0]}), (s2:Synset {id: rel[1]})
    CALL apoc.create.relationship(s1, rel[2], {}, s2) YIELD rel AS newRel
    RETURN newRel
    """
    tx.run(query_relationships, relationships=relationships_data)

    query_examples = """
    UNWIND $examples AS example
    MERGE (e:Example {text: example.text})
    WITH e, example.synset_id AS synset_id  // Keep synset_id for MATCH
    MATCH (s:Synset {id: synset_id})
    MERGE (s)-[:EXAMPLE]->(e)
    """
    tx.run(query_examples, examples=examples_data)
```

```python
def insert_wordnet_data_batch():

    batch_size = 50
    total_batches = len(all_synsets) // batch_size + (1 if len(all_synsets) % batch_size != 0 else 0)

    for i in range(0, len(all_synsets), batch_size):
        batch = all_synsets[i:i + batch_size]
        print(f"Processing batch {i // batch_size + 1} of {total_batches}...")
        process_synset_batch(batch)
```
